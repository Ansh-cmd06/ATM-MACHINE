```
plt.figure(figsize=(8,5))
sns.histplot(df['Tenure Months'],bins = 30,kde = True)
plt.xlabel('Tenure Months')
plt.ylabel('Customer count')
plt.title('distribution of tenure months vs no. of customer')
plt.show()
```
